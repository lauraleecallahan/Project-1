```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pathlib import Path
import csv
```

```python
# File to Load 
gaming_men_to_load = "Data/res.csv"
gaming_anx_to_load = "Data/GamingStudy_data_try2.csv"
```

```python
# Read Gaming Data Files and store into Pandas DataFrames
gaming_men = pd.read_csv(gaming_men_to_load)
gaming_anx = pd.read_csv(gaming_anx_to_load)
```

```python
gaming_anx.head()
```

#### Clean the Data

```python
# editing timestamp to show date
gaming_anx['Timestamp'] = pd.to_datetime(gaming_anx['Timestamp'], origin='1899-12-30', unit='D')
gaming_anx.head()
```

```python
# dropping highestleague column due to all null results
gaming_anx = gaming_anx.drop('highestleague', axis=1)
print(gaming_anx.columns)
```

```python
# Remove rows with any missing values
rows_before = gaming_anx.shape[0]
df_cleaned_no_missing = gaming_anx.dropna()
rows_after = df_cleaned_no_missing.shape[0]

# Display results
rows_removed = rows_before - rows_after
df_cleaned_no_missing.info(), f"Rows removed: {rows_removed}"
```

```python
# Attempt to read the new file and create GAD_Sum and SWL_Sum columns
try:
    # Add a "GAD_Sum" column by summing the specified columns
    df_cleaned_no_missing['GAD_Sum'] = df_cleaned_no_missing[['GAD1', 'GAD2', 'GAD3', 'GAD4', 'GAD5', 'GAD6', 'GAD7']].sum(axis=1)
    # Add a "SWL_Sum" column by summing the specified columns
    df_cleaned_no_missing['SWL_Sum'] = df_cleaned_no_missing[['SWL1', 'SWL2', 'SWL3', 'SWL4', 'SWL5']].sum(axis=1)
    # Display the updated DataFrame structure
    df_cleaned_no_missing[['GAD_Sum', 'SWL_Sum']].head(), df_cleaned_no_missing.info()
except Exception as e:
    e
```

```python
print(df_cleaned_no_missing.columns)
```

```python
# Reordering columns using loc
df_cleaned_no_missing = df_cleaned_no_missing.loc[:, ['S. No.', 'Timestamp', 'GAD1', 'GAD2', 'GAD3', 'GAD4', 'GAD5', 'GAD6',
       'GAD7', 'GADE', 'GAD_Sum', 'SWL1', 'SWL2', 'SWL3', 'SWL4', 'SWL5','SWL_Sum', 'Platform',
       'Hours', 'earnings', 'whyplay', 'League', 'streams', 'SPIN1', 'SPIN2',
       'SPIN3', 'SPIN4', 'SPIN5', 'SPIN6', 'SPIN7', 'SPIN8', 'SPIN9', 'SPIN10',
       'SPIN11', 'SPIN12', 'SPIN13', 'SPIN14', 'SPIN15', 'SPIN16', 'SPIN17',
       'Narcissism', 'Gender', 'Age', 'Work', 'Playstyle', 'accept', 'GAD_T',
       'SWL_T', 'SPIN_T']]

print(df_cleaned_no_missing.columns)
```

#### How does game time affect Mental Well-Being?

```python
# Importing libraries 
import matplotlib.pyplot as plt
import seaborn as sns

# Define bins for Hours and GAD_Sum
hours_bins_5_limited_100 = np.arange(0, 105, 5)  # Hours in increments of 5 up to 100
gad_bins_1 = np.arange(0, 22, 1)  # GAD_Sum in increments of 1

# Create the 2D histogram
heatmap_data_y_1_new, x_edges_5_limited, y_edges_1_new = np.histogram2d(
    df_cleaned_no_missing['Hours'], df_cleaned_no_missing['GAD_Sum'], bins=[hours_bins_5_limited_100, gad_bins_1]
)

# Plot the heatmap
plt.figure(figsize=(16, 10))  # Full-screen dimensions
sns.heatmap(
    heatmap_data_y_1_new.T[::-1],  # Flip the heatmap vertically
    xticklabels=np.round(hours_bins_5_limited_100[:-1], 1),
    yticklabels=gad_bins_1[::-1],  # Flip the GAD_Sum labels
    cmap="Blues",
    cbar_kws={'label': 'Count'},
    annot=False  # Remove annotations
)

# Adjusting labels and titles
plt.title('Heatmap of GAD_Sum vs Hours (Y-Binned in 1s, X-Limit at 100)', fontsize=18)
plt.xlabel('Hours (Binned in 5s)', fontsize=14)
plt.ylabel('GAD_Sum (Descending, Binned in 1s)', fontsize=14)
plt.tight_layout()  # Ensure layout fits well
plt.show()
```

#### How does game play affect life satisfaction?

```python

# Define relevant data
swl_hours_data = df_cleaned_no_missing[['SWL_Sum', 'Hours']].dropna()

# Define bins for Hours and SWL_Sum
hours_bins_5_limited_100 = np.arange(0, 105, 5)  # Hours in increments of 5, up to 100
swl_bins_1 = np.arange(0, swl_hours_data['SWL_Sum'].max() + 2, 1)  # SWL_Sum in increments of 1

# Create the 2D histogram for SWL_Sum and Hours
heatmap_data_swl, x_edges_swl, y_edges_swl = np.histogram2d(
    swl_hours_data['Hours'], swl_hours_data['SWL_Sum'], bins=[hours_bins_5_limited_100, swl_bins_1]
)

# Plot the heatmap with the y-axis starting at 6 and ending at 35
plt.figure(figsize=(16, 10))  # Full-screen dimensions
sns.heatmap(
    heatmap_data_swl.T,  # Do not flip vertically for ascending order
    xticklabels=np.round(hours_bins_5_limited_100[:-1], 1),
    yticklabels=swl_bins_1[0:36],  # Ensure labels from 6 to 35
    cmap="Blues",
    cbar_kws={'label': 'Count'},
    annot=False  # Remove annotations
)

# Adjusting the y-axis to include only values from 6 to 35
plt.ylim(6, 36)  # Limit y-axis to 6 to 35
plt.title('Heatmap of SWL_Sum vs Hours (Y-Binned in 1s, X-Limit at 100, Y-Ascending 6-35)', fontsize=18)
plt.xlabel('Hours (Binned in 5s)', fontsize=14)
plt.ylabel('SWL_Sum (Ascending, Binned in 1s)', fontsize=14)
plt.tight_layout()  # Ensure layout fits well
plt.show()
```
